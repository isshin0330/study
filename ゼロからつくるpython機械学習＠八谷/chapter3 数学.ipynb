### L2ノルム：同じベクトル同士の内積の平方根
$$
    \boldsymbol{w} = 
    \left[\begin{array}{c}
        2 \\
        3 \\
    \end{array}\right] \quad
    \boldsymbol{w^{\mathrm{T}}} = 
    \left[\begin{array}{c}
        2 & 3 \\
    \end{array}\right]\\
    ||\boldsymbol{w}|| = 
    \sqrt{\boldsymbol{w^{\mathrm{T}}}\boldsymbol{w}} = 
    \sqrt{2^2+3^2}
$$

### P.56 ~~ベクトルxとベクトルwの内積xw=||x||・||w||cosθはコサイン類似度とも呼ばれ、2つのベクトルの類似度を表す。~~
### cosθの値をコサイン類似度と呼ぶ。－１≦θ≦１

```python
#コサイン類似度：cosine_similarityを定義
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
w = np.array([[5.0], [3.0]])
#print(np.linalg.norm(w, ord=0))
print(np.linalg.norm(w, ord=1))
print(np.linalg.norm(w, ord=2))
```

```python
w = np.array([[5.0], [3.0]])
x = np.array([1.0, 5.0])
xw = np.matmul(x, w)
ww = np.matmul(w.T, w)

print(f"xとwの内積)　形：{xw.shape}, 型：{xw.dtype}\n{xw}\n")
print(f"wのノルム)　形：{ww.shape}, 型：{ww.dtype}\n{ww}\n")
print(cos_sim(w.T, x))
```

## 逆行列

```python
A = np.array([ [6, 2], [2, 5] ])
B = np.array([ [6, 3], [2, 1] ])

rankA = np.linalg.matrix_rank(A)
rankB = np.linalg.matrix_rank(B)

if rankA == len(A):
    invA = np.linalg.inv(A)
    print(f"行列A) ランク:{rankA}\n 逆行列:\n{invA}\n")
else:
    print(f"行列A) ランク:{rankA}, 特異行列\n")
    
if rankB == len(B):
    invB = np.linalg.inv(B)
    print(f"行列B) ランク:{rankB}\n 逆行列:\n{invB}\n")
else:
    print(f"行列B) ランク:{rankB}, 特異行列\n")
```

# 固有値(eigenvalues)・固有ベクトル(eigenvectors)の性質

```python
import numpy as np
A = np.array([ [3,1], [1,3] ])
L, V = np.linalg.eig(A)

print(f"行列Aの固有値：\n{L}\n")

print(f"行列Aの固有ベクトル：\n{V}\n")

print(f"固有ベクトルの内積：{np.matmul(V[:, [0]].T, V[:, [1]])}\n")

print(f"固有ベクトルのドット積：{np.dot(V[:, [0]].T, V[:, [1]])}\n")

print(f"固有値の和：{np.sum(L)}\n")

print(f"行列Aのトレース（対角成分の和）：{np.sum(np.diag(A))}\n")

print(f"固有値の積：{np.prod(L)}\n")

print(f"行列Aの行列式：{np.linalg.det(A):.1f}\n")

print(f"行列Vの逆行列：\n{np.linalg.inv(V)}\n")

print(f"行列Vの転置：\n{V.T}\n")
```

## ベルヌーイ分布を書いてみる

```python
import numpy as np
from scipy.stats import binom

# サイコロを10回振って、1の目が3回出る確率
p3 = binom.pmf(3, 10, 1/6)

import matplotlib.pyplot as plt

# FigureとAxes
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("k", fontsize = 16)
ax.set_ylabel("P(X=k)", fontsize = 16)

# 分布を決める定数
n = 1
p = 7 / 13
k = np.arange(0, 2)

# P(X=k)を計算
pk = binom.pmf(k, n, p) #試行は一回なので青がでるのは０or１　ゆえにk=0 , 1

# 二項分布をプロット
plt.ylim(0,1)
color_list = ["b", "r"]
ax.bar(k, pk, color = color_list, ec="black")
```

## 三次元グラフ描画

```python
#3次元グラフ描画練習
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def func1(x, y):
    return x**2 + y**2

x = np.arange(-3.0, 3.0, 0.1)
y = np.arange(-3.0, 3.0, 0.1)

X, Y = np.meshgrid(x, y)
Z = func1(X, Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

ax.plot_wireframe(X, Y, Z)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math

#情報エントロピーの概念理解のため、情報が三次元だった場合のグラフを描画してみる
def func1(x, y):
    return -x*np.log(x) - y*np.log(y) - (1-x-y)*np.log(1-x-y)

x = np.arange(0, 1, 0.1)
y = np.arange(0, 1, 0.1)

X, Y = np.meshgrid(x, y)
Z = func1(X, Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

ax.plot_wireframe(X, Y, Z)
plt.show()
```
